```
!pip install selenium
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
from selenium import webdriver
```

```
url = 'https://www.investing.com/equities/trending-stocks'
```

```
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, "html.parser")
```

```
trending_stocks_soup = soup.findAll('div', id='microChartData')
```

```
trending_stocks = [(stock.findAll('a')[0].text, stock.findAll('div')[0].findAll('div')[0].text) for stock in trending_stocks_soup]

print(trending_stocks)
```

```
page = wd.page_source
page_soup = BeautifulSoup(page,'html.parser')
```

```
charts = page_soup.findAll('g', {'class': 'highcharts-xaxis-labels'})
charts = [chart.findAll('text') for chart in charts]
charts = [chart for chart in charts if chart ]
charts = [[text.text for text in chart] for chart in charts]
charts
```
